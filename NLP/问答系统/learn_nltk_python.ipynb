## 停用词

```python
example_sentence = "This is an example showing off stop word filtration"
stop_words = set(stopwords.words("english"))
stop_words
```

```python
words = word_tokenize(example_sentence)
words
```

```python
filtered_sentence = []
for w in words:
    if w not in stop_words:
        filtered_sentence.append(w)
filtered_sentence
```

## 词性标注

```python
from nltk.corpus import state_union
from nltk.tokenize import PunktSentenceTokenizer #默认句子标记化器的抽象类

train_text = state_union.raw("2005-GWBush.txt")
sample_text = state_union.raw("2006-GWBush.txt")
print(sample_text)
custom_sent_tokenizer = PunktSentenceTokenizer(train_text)
custom_sent_tokenizer
```

```python
tokenized = custom_sent_tokenizer.tokenize(sample_text)
tokenized
```

```python
def process_content():
    try:
        for i in tokenized:
            words = nltk.word_tokenize(i)
            tagged = nltk.pos_tag(words)
            print(tagged)
    except Exception as e:
        print(str(e))
    
process_content()
```

```python
# 正则表达式过滤
import re
emoticons_str = r"""
    (?:
        [:=;]
    )
"""

emoticons_str = r"""
    (?:
        [:=;] # 眼睛
        [oO\-]? # ⿐⼦
        [D\)\]\(\]/\\OpP] # 嘴
    )"""
regex_str = [
    emoticons_str,
    r'<[^>]+>', # HTML tags
    r'(?:@[\w_]+)', # @某⼈
    r"(?:\#+[\w_]+[\w\'_\-]*[\w_]+)", # 话题标签
    r'http[s]?://(?:[a-z]|[0-9]|[$-_@.&+]|[!*\(\),]|(?:%[0-9a-f][0-9a-f]))+',
                   # URLs
    r'(?:(?:\d+,?)+(?:\.?\d+)?)', # 数字
    r"(?:[a-z][a-z'\-_]+[a-z])", # 含有 - 和 ‘ 的单词
    r'(?:[\w_]+)', # 其他
    r'(?:\S)' # 其他
]

tokens_re = re.compile(r'('+'|'.join(regex_str)+')', re.VERBOSE | re.IGNORECASE)
emoticon_re = re.compile(r'^'+emoticons_str+'$', re.VERBOSE | re.IGNORECASE)
  
def tokenize(s):
    return tokens_re.findall(s)
  
def preprocess(s, lowercase=False):
    tokens = tokenize(s)
    if lowercase:
        tokens = [token if emoticon_re.search(token) else token.lower() for token in tokens]
    return tokens
  
tweet = 'RT @angelababy: love you baby! :D http://ah.love #168cm'
print(preprocess(tweet))
```

## 闭包：一个返回值是函数的函数
 调用后打印当前时间

```python
import time
def runtime():
    def now_time():
        print(time.time())
    return now_time

f = runtime()
f()
```

## 装饰器、语法糖、注解

```python
# 获取函数运行时间
def runtime(func):
    def get_time():
        print(time.time())
        func()
    return get_time

@runtime
def student_run():
    print("学生跑")
    
student_run()
```

```python
# 有参数的装饰器
def runtime(func):
    def get_time(*args):
        print(time.time())
        func(*args)
    return get_time

@runtime
def student_run(i,u):
    print("学生跑")
    
student_run(1,2)
```

```python
# 关键字参数， key=value
def runtime(func):
    def get_time(*args,**kwargs):
        print(time.time())
        func(*args,**kwargs)
    return get_time

@runtime
def student_run(*args,**kwargs):
    print("学生跑")
    
@runtime
def student_run2(*args,**kwargs):
    print("学生跑2")
    
student_run(3, i = 1,u = 2)

student_run2()
```
