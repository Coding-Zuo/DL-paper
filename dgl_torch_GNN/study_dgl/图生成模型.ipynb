# 如何同时训练和生成一个图

探索图像嵌入过程中进行并行操作，这是不可少的构建模块。

更改Deep Graph Library（DGL）中的图形的基础操作无非就是add_nodes和add_edges。

```python
import dgl

g = dgl.DGLGraph()
g.add_nodes(1) #添加编号为0的节点
g.add_nodes(1) #添加编号为1的节点

# 默认情况下，DGLGraph中的边是有向的
# 对于无方向的边，为两个方向都添加边
g.add_edges([1,0],[0,1]) # 添加由节点1到节点0的边 来回
g.add_nodes(1) #添加编号为2的节点
g.add_edges([2,1],[1,2]) # 添加由节点2到节点1的边 来回
g.add_edges([2, 0], [0, 2]) # 添加由节点2到节点0的边 来回
```

图具有不同的种类，不同种类的图具有不同的大小，拓扑，节点类型，边类型并且可能是多重图。<br>
无论如何，生成过程需要以下几步：
- 对变换的图像进行编码
- 随机执行操作
- 在训练过程中，收集错误信号并优化模型参数
