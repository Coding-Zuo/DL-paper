# 用numpy实现两层神经网络
一个全连接relu神经网络，一个隐藏层，没有bias。用来从x预测y，使用l2 loss

- h = W_1X
- a = max(0,h)
- y_hat = W_2a

- forward pass
- loss
- backward pass

```python
import numpy as np

N=64 # 64个输入
D_in=1000 #输入1000维
H=100 # 中间层 100维
D_out=10 #输出10维
```

```python
# 随机创建一些训练数据
x = np.random.randn(N,D_in)
y = np.random.randn(N,D_out)

w1 = np.random.randn(D_in,H)
w2 = np.random.randn(H,D_out)

learning_rate=1e-6
```

- y:64x10
- x:64x1000

- h =64x1000 dot 1000x100 = 64x100
- h_relu = 64x100
- y_pred = 64x100 dot 100x10 = 64x10
- $loss = \sum(y_{pred}-y)^2$ = 64x10
- $\frac {\partial y_{pred}}{\partial w_2} = h_{relu} , \partial w_2 = h_{relu}^T * \partial y_{pred}$  这里好像是逆才对啊？
- $\frac {\partial y_{pred}}{\partial h_{relu}} = w_2 , \partial h_{relu} = w_2^T * \partial y_{pred}$ 
- $\frac {\partial h}{\partial w_1} = x , \partial w_1 = x^T * \partial h$ 

```python
for it in range(500): 
    # forward pass
    h=x.dot(w1) # N*H
    h_relu = np.maximum(h,0) # N*H
    y_pred = h_relu.dot(w2)  # N*D_out
    
    #compute loss
    loss = np.square(y_pred - y).sum()  # sum和平均只是乘以64的区别
    print(it,loss)
    
    # backword pass
    # cumpute the gradient
    # d loss/d w1 = 链式。。。
    grad_y_pred =2.0*(y_pred-y)
    grad_w2 = h_relu.T.dot(grad_y_pred)
    grad_h_relu = grad_y_pred.dot(w2.T)
    grad_h = grad_h_relu.copy()
    grad_h[h<0]=0
    grad_w1 = x.T.dot(grad_h)
    
    # update weights of w1 and w2
    w1 -= learning_rate*grad_w1
    w2 -= learning_rate*grad_w2
```
